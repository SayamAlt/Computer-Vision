```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/internal_external.png',0)
plt.imshow(img,cmap='gray');
```

```python
contours, hierarchy = cv2.findContours(image=img,mode=cv2.RETR_CCOMP,method=cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_contours = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
    elif hierarchy[0][i][3] != -1:
        cv2.drawContours(external_contours,contours,i,0,-1)
```

```python
plt.imshow(external_contours,cmap='gray');
```

```python
internal_contours = np.zeros(img.shape)
external_contours = np.zeros(img.shape)

for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
    elif hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours,contours,i,255,-1)

plt.subplot(121)
plt.imshow(external_contours,cmap='gray')
plt.title('EXTERNAL CONTOURS')

plt.subplot(122)
plt.imshow(internal_contours,cmap='gray')
plt.title('INTERNAL CONTOURS')

plt.tight_layout()
plt.show();
```
