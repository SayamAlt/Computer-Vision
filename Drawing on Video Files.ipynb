```python
import cv2

video_capture = cv2.VideoCapture(0)

width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# TOP LEFT CORNER OF RECTANGLE
x = width // 4
y = height // 4

# WIDTH AND HEIGHT OF RECTANGLE
w = width // 2
h = height // 2

# BOTTOM RIGHT CORNER OF RECTANGLE: (x+w,y+h)

while True:
    ret, frame = video_capture.read()
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=3)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```

```python
import cv2

video_capture = cv2.VideoCapture(0)

# DEFINE THE CALLBACK FUNCTION
def draw_rectangle(event,x,y,flags,params):
    global leftPoint, rightPoint, topLeftPointClicked, bottomRightPointClicked
    
    if event == cv2.EVENT_LBUTTONDOWN:
        
        # RESET THE RECTANGLE IF IT ALREADY EXISTS
        if topLeftPointClicked and bottomRightPointClicked:
            leftPoint = (0,0)
            rightPoint = (0,0)
            topLeftPointClicked = False
            bottomRightPointClicked = False
        
        if topLeftPointClicked == False:
            leftPoint = (x,y)
            topLeftPointClicked = True
        
        elif bottomRightPointClicked == False:
            rightPoint = (x,y)
            bottomRightPointClicked = True
        
# INITIALIZE THE GLOBAL VARIABLES
leftPoint = (0,0)
rightPoint = (0,0)
topLeftPointClicked = False
bottomRightPointClicked = False

# SET CALLBACK TO THE WINDOW 
cv2.namedWindow('TEST')
cv2.setMouseCallback('TEST',draw_rectangle)

while True:
    ret, frame = video_capture.read()
    
    if topLeftPointClicked:
        cv2.circle(frame,center=leftPoint,radius=5,color=(0,0,255),thickness=-1)
    
    if topLeftPointClicked and bottomRightPointClicked:
        cv2.rectangle(frame,leftPoint,rightPoint,color=(255,0,128),thickness=3)
        
    cv2.imshow('TEST',frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
