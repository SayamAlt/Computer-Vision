```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
denis = cv2.imread('../DATA/Denis_Mukwege.jpg',0)
nadia = cv2.imread('../DATA/Nadia_Murad.jpg',0)
solvay = cv2.imread('../DATA/solvay_conference.jpg',0)
```

```python
def display(image):
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap='gray')
```

```python
display(denis);
```

```python
display(nadia);
```

```python
display(solvay);
```

```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
face_cascade
```

```python
def detectFace(image):
    img = image.copy()
    rectangle = face_cascade.detectMultiScale(img)
    
    for (x,y,w,h) in rectangle:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return img
```

```python
display(detectFace(denis));
```

```python
display(detectFace(nadia));
```

```python
display(detectFace(solvay));
```

```python
def adj_detect_face(image):
    img = image.copy()
    res = face_cascade.detectMultiScale(img,scaleFactor=1.2,minNeighbors=3)
    
    for (x,y,w,h) in res:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return img
```

```python
display(adj_detect_face(solvay));
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read(0)
    cv2.imshow('Video Face Detection',detectFace(frame))
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')
eye_cascade
```

```python
def detectEyes(image):
    img = image.copy()
    res = eye_cascade.detectMultiScale(img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in res:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),10)
    
    return img
```

```python
display(detectEyes(nadia));
```
